# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Lets get the data

```python
df = pd.read_csv('imdb_top_1000.csv')

df.head()
```

# Lets answer the questions in the notebook

### 1. What are the column names in the dataset?

```python
cols = df.columns
cols
```

### 2. What is the average rating of the movies?

```python
average_rating = df['IMDB_Rating'].mean()

average_rating
```

### 3. How many unique genres are there?

```python
unique_genres = df['Genre'].unique()
unique_genres_count = len(unique_genres)

unique_genres_count
```

### 4. Save your results as a csv file.

```python
# I create a new df containing the questions and answers for the analysis
df_analysis = pd.DataFrame({
    'Question': ['What are the column names in the dataset?', 'What is the average rating of the movies?', 'How many unique genres are there?'],
    'Answer': [list(cols), average_rating, unique_genres_count]
})
```

```python
df_analysis
```

```python
# save the df to a csv file
df_analysis.to_csv('analysis.csv', index=False)
```

### 5. Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list.

```python
# plot histogram for genre distribution
plt.hist(df['Genre'], bins=20)
plt.xticks(rotation=90)
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

### 6. Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list?)

```python
counts = df['Released_Year'].value_counts()
plt.xticks(rotation=90)
plt.scatter(counts.index, counts.values)
```
